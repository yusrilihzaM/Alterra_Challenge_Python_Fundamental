```
#Import library
import math as mt
```

```
#Soal 1
#Input jari-jari dan tinggi 
r=float(input("Masukan Jari-jari= "))
t=float(input("Masukan Tinggi= "))
#phi menggunakan math.pi
phi=mt.pi
#Perhitungan
Lp=2*phi*r**2+2*phi*r*t
#Output
print("==============================")
print("Output:",Lp)
```

```
#input->pengecekan->tampilkan
#inputan
angka = int(input("Masukkan bilangan= "))
#pengecekan->tampilkan
#if parent
:#bilangan harus bukan dari 1
if angka > 1
  for i in range(2,angka):
    #if child
    if (angka % i) == 0:
      print("Output:")
      print(angka, "bukan bilangan prima")
      break
  else:
    print("Output:")
    print(angka,"adalah bilangan prima")
    #endif child
else:
  print("Output:")
  print(angka, "bukan bilangan prima")
#endif parent
```

```
#input->pengecekan kata balik=kata ori->tampilkan
#input
kata=str(input("Masukan Kata:"))
#pengecekan
if(kata==kata[::-1]):
  print("Output: ")
  print(kata+" Adalah palindrome")
else:
  print("Output: ")
  print(kata+" Bukan palindrome")
```
